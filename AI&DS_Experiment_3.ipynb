```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/apple_quality.csv')
```

```python
df.describe()
```

```python
df.info()
```

```python
#check for missing values
print(df.isnull().sum())
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df['Quality'].value_counts()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Quality'] = le.fit_transform(df['Quality'])
df.head()
```

```python
df = df.drop(columns = ['A_id'])
df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns = ['Quality'])
Y = df['Quality']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25)
```

```python
plt.figure(figsize=(8, 6))
plt.bar(
    ["Training set", "Testing set"],
    [Y_train.value_counts()[1], Y_test.value_counts()[1]],
    color=["skyblue", "lightgreen"],
)
plt.xlabel("Dataset")
plt.ylabel("Count of quality 1")
plt.title("Distribution of quality 1 in training and testing sets")
plt.show()

# Visualize the distribution of features in training and testing sets (assuming numerical features)
for col in X_train.columns:
    plt.figure(figsize=(8, 6))
    plt.hist(X_train[col], bins=10, alpha=0.5, label="Training set")
    plt.hist(X_test[col], bins=10, alpha=0.5, label="Testing set")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.title(f"Distribution of feature '{col}' in training and testing sets")
    plt.legend()
    plt.show()
```

```python
from statsmodels.stats.weightstats import ztest as ztest
ztest(Y_test, Y_train, value=0)
```

```python
import pandas as pd
from scipy.stats import ttest_ind

def result(p):
    if p['p_value'] < 0.05:
        p['hypothesis_accepted'] = 'alternative'
    else:
        p['hypothesis_accepted'] = 'null'

    df = pd.DataFrame(p, index=[''])
    cols = ['value1', 'value2', 'score', 'p_value', 'hypothesis_accepted']
    return df[cols]

p = {}
p['value1'], p['value2'] = Y_train.mean(), Y_test.mean()
p['score'], p['p_value'] = ttest_ind(Y_train, Y_test, alternative='two-sided')

result(p)
```
