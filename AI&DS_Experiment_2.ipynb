```python
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
countries = ["Finland", "Denmark", "Iceland", "Switzerland", "Netherlands",
             "Burundi", "Central African Republic", "South Sudan", "Chad", "Yemen"]
ladder_scores = [7.842, 7.620, 7.557, 7.509, 7.449, 2.325, 2.291, 2.284, 2.232, 2.029]

# Split data into top and bottom 5 countries
top_5_countries = countries[:5]
top_5_scores = ladder_scores[:5]

bottom_5_countries = countries[-5:]
bottom_5_scores = ladder_scores[-5:]

# Create a bar graph with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Top 5 countries
ax1.bar(top_5_countries, top_5_scores, color="skyblue", label="Top 5")
ax1.set_xlabel("Country")
ax1.set_ylabel("Ladder Score")
ax1.set_title("Top 5 Happiest Countries")
ax1.legend()

# Bottom 5 countries
ax2.bar(bottom_5_countries, bottom_5_scores, color="coral", label="Bottom 5")
ax2.set_xlabel("Country")
ax2.set_ylabel("Ladder Score")
ax2.set_title("Bottom 5 Least Happy Countries")
ax2.legend()

# Adjust spacing between subplots
plt.subplots_adjust(wspace=0.3)

plt.show()
```

```python
df_india = df[df['Country name'] == 'India']
regional_mean_scores = df.groupby('Regional indicator')['Ladder score'].mean().reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(x='Regional indicator', y='Ladder score', data=regional_mean_scores, color='skyblue', label='Average Ladder Score')
sns.barplot(x=['India'], y=df_india['Ladder score'].iloc[0], color='orange', label='India')
plt.title('Average Ladder Score by Regional Indicator (with India Comparison)')
plt.xlabel('Regional Indicator')
plt.ylabel('Average Ladder Score')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
regional_mean_corruption = df.groupby('Regional indicator')['Perceptions of corruption'].mean().reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(x='Regional indicator', y='Perceptions of corruption', data=regional_mean_corruption, color='skyblue')
plt.title('Average Perceptions of Corruption by Regional Indicator')
plt.xlabel('Regional Indicator')
plt.ylabel('Average Perceptions of Corruption')
plt.xticks(rotation=45)
plt.show()
```

```python
# Identify outliers using IQR method
Q1 = df['Ladder score'].quantile(0.25)
Q3 = df['Ladder score'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Ladder score'] < lower_bound) | (df['Ladder score'] > upper_bound)]

# Remove outliers
df_no_outliers = df[~((df['Ladder score'] < lower_bound) | (df['Ladder score'] > upper_bound))]
print("Number of outliers removed:", outliers.shape[0])

# Updated box plot without outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x='Regional indicator', y='Ladder score', data=df_no_outliers)
plt.xticks(rotation=45)
plt.title('Box Plot of Ladder Score by Regional Indicator (without outliers)')
plt.xlabel('Regional Indicator')
plt.ylabel('Ladder Score')
plt.show()
```
